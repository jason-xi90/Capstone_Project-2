```python
Under20 = players22[players22['age'] <= 20]
age20_25 = players22[(players22['age']>20)&(players22['age']<=25)]
age25_30 = players22[(players22['age']>25)&(players22['age']<=30)]
Above30 = players22[players22['age']>30]
```

```python
x = np.array([Under20['short_name'].count(),age20_25['short_name'].count(),age25_30['short_name'].count(),Above30['short_name'].count()]) 
mylabels = ["<=20",">20&<=25",">25&<=30",">=30"]
plt.title('Total Players with Age Groups',fontsize=20)
plt.pie(x,labels=mylabels,autopct="%.1f%%")
plt.show()
```

```python
players_under_20 = players22[players22['age']<20]
players_under_20['club_name'].value_counts().plot(kind = 'bar',color=sns.color_palette("cubehelix"))
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='club_name', y='age', data=players22)
plt.xticks(rotation=90)
```

```python
num_player = players22.groupby('club_name').size()
data=(players22.groupby('club_name')['age'].sum())/num_player
data.sort_values(ascending = False)
```

```python
wage_by_clubs = pd.DataFrame(players22.groupby('club_name', as_index=False)['wage_eur'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax=sns.barplot(x='club_name',y='wage_eur',data=wage_by_clubs.sort_values(by='wage_eur'),palette='Set2')
ax.set_xlabel("club_name", fontsize=30)
ax.set_ylabel("wage_eur", fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Wage', fontsize=20)
```
